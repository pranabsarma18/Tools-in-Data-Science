## Q14: Predict sentiment using Textblob. How many positive titles (title) are there [books.csv] ? (cut-off >0) N

```python
from textblob import TextBlob
```

```python
books1 = books.copy()
```

```python
books1.head()
```

```python
books1['TextBlob_Subjectivity'] = books1['title'].apply(lambda x: TextBlob(x).sentiment.subjectivity)
books1['TextBlob_Polarity'] = books1['title'].apply(lambda x: TextBlob(x).sentiment.polarity)
```

```python
books1.title
```

```python
books1.head()
```

```python
(books1["TextBlob_Polarity"] > 0).sum()
```
